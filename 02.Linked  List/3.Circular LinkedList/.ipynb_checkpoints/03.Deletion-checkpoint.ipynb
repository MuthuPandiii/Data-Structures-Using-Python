# Deletion

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Circular_LinkedList:
    def __init__(self):
        self.last=None
    
    def add_at_empty_list(self,data):
        if self.last!=None:
            return self.last
        new_node=Node(data)
        self.last=new_node
        self.last.next=self.last
        return self.last
        
    def add_at_end(self,data):
        if self.last==None:
            return self.add_at_empty_list(data)
        new_node=Node(data)
        new_node.next=self.last.next
        self.last.next=new_node
        self.last=new_node
        return self.last
    
    def print_list(self):
        if self.last==None:
            print("The list is Empty")
            return
        temp=self.last.next
        while temp:
            print(temp.data,end=" ")
            temp=temp.next
            if temp==self.last.next:
                break
    
    def delete_by_data(self,key):
        if self.last.data==key and self.last.next==self.last:
            self.last=None
            return self.last
        elif  self.last.next.data==key:
            head=self.last.next
            temp=head
            while temp.next!=self.last.next:
                temp=temp.next
            temp.next=head.next
            return self.last
        else:
            head=self.last.next
            curr=head
            prev=head
            while curr.next!=head:
                if curr.data==key:
                    prev.next=curr.next
                    curr=None
                    if prev.next==head:
                        self.last=prev
                        return self.last
                    else:
                        return self.last
                prev=curr
                curr=curr.next
            print("\nThe item not in the List")
            

c=Circular_LinkedList()
c.add_at_end(9)
c.add_at_end(8)
c.add_at_end(7)
c.add_at_end(6)
c.add_at_end(5)
c.print_list()
c.delete_by_data(7)
print("\nAfter Deleting 7")
c.print_list()
c.delete_by_data(8)
print("\nAfter Deleting 8")
c.print_list()
c.delete_by_data(9)
print("\nAfter Deleting 9")
c.print_list()
c.delete_by_data(6)
print("\nAfter Deleting 6")
c.print_list()
c.delete_by_data(5)
print("\nAfter Deleting 5")
c.print_list()

print("\n\n")
c=Circular_LinkedList()
c.add_at_end(9)
c.add_at_end(8)
c.add_at_end(7)
c.add_at_end(6)
c.add_at_end(5)
c.print_list()

c.delete_by_data(5)
print("\nAfter Deleting 5")
c.print_list()
```
