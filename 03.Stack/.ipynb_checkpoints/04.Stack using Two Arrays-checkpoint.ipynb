```python
class Stack:
    
    def __init__(self,n):
        self.n=n
        self.array=[None]*n
        self.top1=n//2
        self.top2=n//2-1
        
    def Push1(self,x1):
        if self.top1>0:
            self.top1-=1
            self.array[self.top1]=x1
        else:
            print("Stack1 Overflow")
    
    def Push2(self,x2):
        if self.top2<self.n-1:
            self.top2+=1
            self.array[self.top2]=x2
        else:
            print("Stack2 OverFlow")
    
    def Pop1(self):
        if self.top1<(self.n//2):
            x=self.array[self.top1]
            self.top1+=1
            return x
        else:
            print("\nStack1 UnderFlow")
    
    def Pop2(self):
        if self.top2>=self.n//2:
            x=self.array[self.top2]
            self.top2-=1
            return x
        else:
            print("\nStack2 Underflow")

            
S=Stack(8)

S.Push1(4)
S.Push1(3)
S.Push1(2)
S.Push1(1)
S.Push2(5)
S.Push2(6)
S.Push2(7)
S.Push2(8)

print(S.array)
print(S.Pop1())
print(S.Pop1())
print(S.Pop1())
print(S.Pop1())
print(S.Pop1())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
```

## Memory Efficient Way

```python
class Stack:
    def __init__(self,n):
        self.n=n
        self.array=[None]*n
        self.top1=-1
        self.top2=n
    
    def Push1(self,x):
        if self.top1==self.top2-1:
            print("Stack1 Overflow")
        else:
            self.top1+=1
            self.array[self.top1]=x
    
    def Push2(self,x):
        if self.top1==self.top2-1:
            print("Stack2 Overflow")
        else:
            self.top2-=1
            self.array[self.top2]=x
    
    def Pop1(self):
        if self.top1>=0:
            x=self.array[self.top1]
            self.top1-=1
            return x
        print("Stack1 Underflow")
    
    def Pop2(self):
        if self.top2<self.n:
            x=self.array[self.top2]
            self.top2+=1
            return x
        print("Stack2 UnderFlow")
            
S=Stack(6)
S.Push1(1)
S.Push1(2)
S.Push1(3)
S.Push2(6)   
S.Push2(5)       
S.Push2(4)  
print(S.array)
print(S.Pop1())
print(S.Pop1())
print(S.Pop1())
print(S.Pop1())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
print(S.Pop2())
print(S.array)
```
