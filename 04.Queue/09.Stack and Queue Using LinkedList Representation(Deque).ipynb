```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None

class Stack_and_Queue:
    def __init__(self):
        self.front=None
        self.rear=None
    
    def  Enqueue(self,x):
        temp=Node(x)
        
        print("\nElement Enqueued : ",x)
        
        if self.rear==None:
            self.front=self.rear=temp
            return
        
        temp.prev=self.rear
        self.rear.next=temp
        self.rear=temp 
    
    def Dequeue(self):
        
        if self.front==None:
            print("\n\tQueue Overflow")
            return
        temp=self.front
        self.front=temp.next
        self.front.prev=None
        print("\nElement Dequeued : ",temp.data)
        
        if self.front==None:
            self.rear=None
    
    def Push(self,y):
        
        temp=Node(y)
        
        print("\nElement Pushed : ",y)
        
        temp.prev=self.rear
        self.rear.next=temp
        self.rear=temp
        
    def  Pop(self):
        
        if self.rear==None:
            print("\n\tStack underflow")
            return
        temp=self.rear
        print("\nElememt Popped : ",temp.data)
        self.rear=self.rear.prev
        self.rear.next=None
    
    def Print_list(self):
        
        temp=self.front
        print("\nThe List :")
        while(temp):
            print(temp.data,end=" ")
            temp=temp.next
        print("\n")
            
       
```

```python
sq=Stack_and_Queue()

sq.Enqueue(2)
sq.Enqueue(5)
sq.Enqueue(6)

sq.Print_list()

sq.Push(7)
sq.Push(8)
sq.Push(9)

sq.Print_list()

sq.Dequeue()
sq.Dequeue()

sq.Print_list()

sq.Pop()
sq.Pop()
sq.Pop()
sq.Pop()
sq.Pop()
```
